### #task 3:

```python
missing_valye = ["N/a","na",np.nan]
df=pd.read_csv(r"D:\salaries\salaries.csv" , na_values =missing_valye )
```

```python
df.isnull().sum()
```

```python
#I dropped Notes and Status because we don't need them
df.drop('Notes', axis=1, inplace=True)
df.drop('Status', axis=1, inplace=True)
```

```python
# filling missing valuby mean 
df['BasePay']=df['BasePay'].fillna(df['BasePay'].mean())

#I filled in the missing values with a value of zero, 
#and since it is an empty value, this means that the employee has no Benefits.
df['Benefits']=df['Benefits'].fillna(0)

# I dropped the rows that have null valus in OvertimePay and OtherPay
df.dropna(subset=['OvertimePay'], inplace=True)
df.dropna(subset=['OtherPay'], inplace=True)

df.isnull().sum()
```

### task 4 :

```python
import matplotlib.pyplot as plt


plt.hist(df['TotalPay'], bins=30, color='green', edgecolor='black') 
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
department_counts = df['JobTitle'].value_counts()
plt.figure(figsize=(20, 20))
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', colors=plt.cm.tab20.colors)
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

### task 5 :

```python
summary_stats = df.groupby(['BasePay', 'OtherPay','OvertimePay'])[['BasePay', 'OvertimePay', 'OtherPay', 'TotalPay']].agg(['mean', 'median', 'std'])
print(summary_stats)
average_salary_comparison = summary_stats['TotalPay']['mean'].reset_index()
print(average_salary_comparison)
```

### task 6 :

```python
correlation = df['TotalPay'].corr(df['Year'])

print("Correlation coefficient:", correlation)

# Creating a scatter plot to visualize the relationship
plt.figure(figsize=(6, 6))
plt.scatter(df['Year'], df['TotalPay'], alpha=1)
plt.title('Scatter Plot of Salary vs Year of Work')
plt.xlabel('Years ')
plt.ylabel('Salary')
plt.grid(True)
years = [2011, 2012, 2013, 2014] 
plt.xticks(years)
plt.show()
```
